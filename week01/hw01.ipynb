```python
tempdf = df.groupby(by='year')['country'].apply(lambda x: x.unique()).reset_index()
```

The above code groups the data by year and applies lambda function to get the unique values in country column. the result is stored in tempdf

```python
countries = {}
for index, row in tempdf.iterrows():
    for country in row["country"]:
        if country in countries:
            countries[country] += 1
        else:
            countries[country] = 1
```

the code creates countries dictionary and stores frequency of each unique country name from tempdf.

```python
tempdf = pd.DataFrame.from_dict(countries, orient='index', columns=['Count'])
tempdf = tempdf.reset_index()
tempdf.columns = ['Country', 'Count']
tempdf
```

```python
tempdf[tempdf['Count'] < 34]
```

By above code when we put condition such that count less than 34 we get the above 9 countries which have missing years.

# Question 2: Wide data isn't usually helpful.
Convert this to a long format with the following columns:

1. country

2. year

3. diagnosis method: mapping should be:

  {'newsp':'Pulmonary Smear (+)'
  
  ,'newsn':'Pulmonary Smear (-)'
  
  ,'newep':'Extrapulmonary'
  
  ,'newrel':'Relapse'}  
  
  
4. gender: male or female.

5. age: lower_age - higher_age, e.g., 0-14

6. number of cases

Note: You can use replace (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html)

```python
newdf = df.reset_index().melt(id_vars=['index', 'year', 'country'], 
                               value_vars=list(df.columns[2:]), 
                               var_name='diagnosis_method', value_name='number of cases')
newdf = newdf.set_index('index')
```

```python
newdf.shape
```

```python
newdf['gender'] = ['Male' if x.split('_')[1][0] == 'm' else 'Female' for x in newdf['diagnosis_method']]
```

```python
newdf['Age'] = newdf['diagnosis_method'].str.extract(r'_(\w+)_')
newdf['Age'] = newdf['Age'].str.replace('m', '0 - 4').replace('n', '5 - 18').replace('o', '18 - 49').replace('p', '50 +')
```

code uses str.extract method to get some part of diagnosis_method. Further the code replaces accordingly the age by given condition

```python
df.rename(columns={col: col.replace('new_s', 'Pulmonary Smear')
                      .replace('new_ep', 'Extrapulmonary')
                      .replace('newrel', 'Relapse') for col in df.columns},
          inplace=True)
```

```python
newdf['diagnosis_method'] = newdf['diagnosis_method'].str.split('_', expand=True)[0]
```

Converting long format to required columns only by using rename and replace.

# Question 3: Create a graph that shows when countries started to report TB cases.

```python
newdf = newdf[~newdf['number of cases'].isna()]
newdf = newdf[newdf['number of cases'] > 0]
newdf['country'].value_counts()
```

the code drops the rows of null values or 0 because we have to find the least number of cases in which year . if 0 number of cases are there it is not possible.

```python
tb = newdf.groupby(by='country').agg({'year': 'min'}).reset_index()
```

```python
tb['year'].value_counts()
```

```python
plt.figure(figsize=(15,15))
plt.ylim(min(tb['year'])-10,max(tb['year'])+10)
ax = sns.lineplot(y=tb['year'], x=tb[:50]['country'])
plt.title("TB Cases in each country with reported years")
plt.xlabel("Country")
plt.ylabel("Year")
ax.tick_params(axis='x', rotation=90)
plt.show()
```

The above line plot shows for each country which year does the TB reporting started above the country name as sharp point like vertex.
